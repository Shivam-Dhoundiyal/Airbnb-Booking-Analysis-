# **4. How do the different room categories vary in Brooklyn and Manhattan in terms of pricing and availability?**

#### Chart - 4

````python
# Set up the subplots
fig, axes = plt.subplots(nrows=1, ncols=3, figsize=(24, 10))
ax = axes.flatten()

# Group by neighbourhood group and room type to get mean price
mean_price_df = df.groupby(['neighbourhood_group', 'room_type'], as_index=False)[['price']].mean()

# Plot mean price by neighbourhood group and room type
sns.barplot(data=mean_price_df, x='neighbourhood_group', y ='price', hue='room_type', ax=ax[0])
ax[0].set_title('Mean Price by Neighbourhood Group and Room Type')

# Plot distribution of prices less than 500 by neighbourhood group
sns.histplot(data=df[df['price'] < 500], x="price", hue="neighbourhood_group", stat='frequency', element='poly', ax=ax[1])
ax[1].set_title('Distribution of Listing Prices (< $500) by Neighbourhood Group')

# Plot distribution of prices by neighbourhood group and room type
sns.histplot(data=df, x='neighbourhood_group', y='price', hue='room_type', element='poly', ax=ax[2])
ax[2].set_title('Distribution of Listing Prices by Neighbourhood Group and Room Type')

# Remove spines from the plots for better readability
sns.despine(fig)
````
